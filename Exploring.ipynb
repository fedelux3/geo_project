## Download NOAA

Step da fare:
1. Applicare la request a tutte le cartelle del sito: https://www.ncei.noaa.gov/data/avhrr-land-normalized-difference-vegetation-index/access/
2. Usare beautifulsoap per prendere tutti i nomi dei file interni
3. Scaricare i raw data direttamente nel disco D

```python
import requests
import os
from bs4 import BeautifulSoup
import urllib
```

```python
data_path = r'D:\Data\geopy\precipitation'
base_link = 'https://www.ncei.noaa.gov/data/global-precipitation-climatology-project-gpcp-monthly/access'
year = 1979
```

```python
for year in range(1980,2021):
    print(f"year {year}")
    page = requests.get(f'{base_link}/{year}')
    fold = f'{data_path}/{year}'
    if not os.path.exists(fold):
        os.makedirs(fold)
    html_doc = page.content.decode('utf-8')
    soup = BeautifulSoup(html_doc, 'html.parser')
    for link in soup.find_all('a'):
        if link.get('href') is None:
            continue
        if link.get('href').endswith('.nc'):
            urllib.request.urlretrieve(f"{base_link}/{year}/{link.get('href')}", f"{data_path}/{year}/{link.get('href')}")
            print(f"{link.get('href')} - DONWLOADED")
```

```python
import xarray as xr
data_path = r'D:\Data\geopy\precipitation'
```

```python
ds_year = []
for year in range(1979,2021):
    for file in os.listdir(f"{data_path}/{year}"):
        ds_year.append(xr.open_dataset(f"{data_path}/{year}/{file}"))
```

```python
len(ds_year)
```

```python
data = xr.concat(ds_year, dim='time')
data
```

```python
#- add cyclic point form plotting purposes
from cartopy.util import add_cyclic_point
import matplotlib as mpl
from matplotlib import pyplot as plt
import cartopy.crs as ccrs # for plot the world
```

```python
pr = data.precip
pr.shape
```

```python
pr_mean = pr.mean(axis=0)
```

```python
#ts_avg_cyc, lon_cyc = add_cyclic_point(ts1.mean(axis=0)-273.15, coord=ts1.lon)  # also Kelvin to Celsius
#norm = mpl.colors.Normalize(vmin=-40, vmax=40) 

# Model IPSL-CM6A-LR
fig = plt.figure(figsize=(9,6))  # x,y(inches)

ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_title('Precipitation - GPCP Monthly\n', size = 14)
ax.set_global()
mm = ax.pcolormesh(pr.longitude, pr.latitude, pr_mean, \
                   transform=ccrs.PlateCarree(),
                   cmap=mpl.cm.Blues)  

ax.coastlines()
ax.gridlines(draw_labels=True)

#- add colorbar
cbar_ax = fig.add_axes([0.28, 0.10, 0.46, 0.05]) #[left, bottom, width, height]
cbar = fig.colorbar(mm, cax=cbar_ax, extend='both', orientation='horizontal')
cbar.set_label('mm')
cbar.ax.tick_params(labelsize=8)

plt.show()
```

```python
base = pr.loc['1979'].mean() # lock one year
```

```python
# spatial and time (yearly) aggragations
pr_space_mean = pr.mean(dim = ['latitude','longitude'])
pr_year_mean = pr_space_mean.groupby('time.year').mean()
pr_year_anomaly = pr_year_mean - base
```

```python
plt.figure(figsize = (11,6))
plt.plot(pr_year_mean['year'], pr_year_anomaly, color = "gray")

plt.title("Compare", size = 17)
#plt.legend(fontsize = 12)
plt.xlabel("time", size = 14)
plt.ylabel("mm", size = 14)
plt.xticks(size = 11)
plt.yticks(size = 11)

plt.grid()
```

```python
pr_month_mean = pr_space_mean.groupby('time.month').mean()
```

```python
months=[]
for month in range(1,13):
    months.append(pr_space_mean[pr_space_mean.time.dt.month == month])
```

```python
plt.subplots(nrows=12, figsize=(8,30))

for i, month in enumerate(months):
    plt.subplot(12,1,i+1)
    plt.plot(month['time'], month, color = "gray")

    plt.title(f"{i+1} Month", size = 17)
    #plt.legend(fontsize = 12)
    plt.xlabel("time", size = 14)
    plt.ylabel("mm", size = 14)
    plt.xticks(size = 11)
    plt.yticks(size = 11)

    plt.grid()
```
